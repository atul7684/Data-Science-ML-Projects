```python
from feature_engine.outliers import Winsorizer

# Lets Winsorize the outlier present in the dataset.

winsor = Winsorizer(capping_method='iqr', tail='both', fold=1.5,
                    variables=['MSSubClass', 'LotFrontage', 'LotArea', 'OverallCond',
                               'YearRemodAdd','MasVnrArea', 'BsmtFinSF1',
                               'BsmtUnfSF', 'TotalBsmtSF', '2ndFlrSF',
                               'GrLivArea','GarageArea', 'WoodDeckSF',
                               'OpenPorchSF'])

# Applying Winsorizer to the selected columns
data[['MSSubClass', 'LotFrontage', 'LotArea', 'OverallCond','YearRemodAdd',
      'MasVnrArea', 'BsmtFinSF1', 'BsmtUnfSF', 'TotalBsmtSF', '2ndFlrSF',
      'GrLivArea', 'GarageArea', 'WoodDeckSF', 'OpenPorchSF']] = \
    winsor.fit_transform(data[['MSSubClass', 'LotFrontage', 'LotArea',
                                'OverallCond','YearRemodAdd','MasVnrArea',
                                'BsmtFinSF1', 'BsmtUnfSF', 'TotalBsmtSF',
                                '2ndFlrSF', 'GrLivArea', 'GarageArea',
                                'WoodDeckSF', 'OpenPorchSF']])
```

Winsorization technique :- Winsorization is a statistical technique used to cap extreme values (outliers) in a dataset. By replacing outliers with a specified quantile (typically the minimum or maximum value within a certain range).

The extreme values are replaced with a specified quantile. For example, if the 5th and 95th percentiles are used, outliers below the 5th percentile are replaced with the 5th percentile value, and outliers above the 95th percentile are replaced with the 95th percentile value.

Key Parameters for Winsorization:

Capping method: The method used to identify outliers (e.g., IQR, standard deviation).

Tail: Whether to cap outliers in both tails (upper and lower) or only one tail.

Fold: The multiplier used to determine the cutoff points for outliers (e.g., a fold of 1.5 means outliers will be capped if they are 1.5 times the IQR away from the interquartile range).
